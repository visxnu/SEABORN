# Histogram

```python
import seaborn as sns
tips=sns.load_dataset('tips')
sns.histplot(x='total_bill',data=tips,kde=True,hue='sex')
```

# KDE Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.kdeplot(tips['total_bill'])
plt.title("KDE plot of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

Horizonatal kde

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.kdeplot(y=tips['total_bill'])
plt.title("KDE plot of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

Curve Color into red

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.kdeplot(tips['total_bill'],color='red')
plt.title("KDE plot of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

Area under curve

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.kdeplot(tips['total_bill'],fill=True)
plt.title("KDE plot of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

Differentiate by size

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.kdeplot(x=tips['total_bill'],hue=tips['sex'],shade=True)
plt.title("KDE plot of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

# Swarm Plot

```python
df=sns.load_dataset('tips')
sns.swarmplot(x='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='tip',data=df)
```

```python
sns.swarmplot(x='day',y='tip',data=df,hue='sex',dodge=True,size=7)
```

```python
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
sns.swarmplot(x='day',y='tip',data=df,color='red')

plt.subplot(1,2,2)
sns.boxplot(x='day',y='tip',data=df,color='green')
```

```python
import matplotlib.pyplot as plt

# plt.subplot(1,2,1)
sns.swarmplot(x='day',y='tip',data=df,hue='sex',dodge=True)

# plt.subplot(1,2,2)
sns.boxplot(x='day',y='tip',data=df,hue='sex')
```
