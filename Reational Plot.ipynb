# Scatter Plot

```python
import seaborn as sns
print(sns.get_dataset_names())
```

```python
df1 = sns.load_dataset("tips")
df1
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df1)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="day",data=df1)
```

```python
sns.scatterplot(x="total_bill",y="tip",style="day",data=df1)
```

```python
sns.scatterplot(x="total_bill",y="tip",size="time",data=df1,sizes=(10,100))
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="sex",
    sizes=(20,200),
    palette="viridis",

    alpha=0.7

)

```

# Line PLot

```python
import seaborn as sns

df = sns.load_dataset("tips")
sns.lineplot(x="size",y="total_bill",color="red",data=df)
```

```python
sns.lineplot(x="size",y="total_bill",hue="day",ci=None,palette="flare",data=df)
```

```python
sns.lineplot(x="size",y="total_bill",style="sex",ci=None,data=df)
```

```python
sns.lineplot(x="size",y="total_bill",size="sex",ci=None,sizes=(1,5),data=df)
```

# Pair Plot

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex")
```

```python
sns.pairplot(tips,hue="sex",markers=["o","s"],diag_kind='hist')
```

```python
sns.pairplot(tips,corner=True)
```

```python
sns.pairplot(tips,height=5,aspect=2)
```

```python
sns.pairplot(tips,vars=["tip","total_bill"])
```
