# Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x="day",palette='rocket',hue='sex',dodge=False,data=tips)
```

# Bar Plot

```python
tips
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(y='day',x='total_bill',data=tips)
```

Control the bar width by customizing figure size

```python
plt.figure(figsize=(20,6))
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar Width Customization")
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator='sum')
plt.show()
```

Adding Hue

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips,palette='pastel',dodge=False)

plt.title("Average Total bill by day and gender")
plt.show()
```

```python
color=['red','black','red','red']

sns.barplot(x='day',y='total_bill',palette=color,data=tips)
```

# Violin Plot

```python
import seaborn as sns
tips=sns.load_dataset('tips')
sns.violinplot(y='day',x='total_bill',data=tips,split=True)
```

```python
import seaborn as sns
tips=sns.load_dataset('tips')
sns.violinplot(y='day',x='total_bill',data=tips,split=True,cut=0)
```
